```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (replace with your actual data)
data = {
 'square_footage': [1500, 2000, 1200, 2500, 1800],
 'bedrooms': [3, 4, 2, 4, 3],
'bathrooms': [2, 2.5, 1.5, 3, 2],
'windows': [10, 12, 8, 14, 11],
'doors': [5, 6, 4, 7, 5],
'flooring': [1, 2, 1, 3, 2],
'house_age': [20, 10, 30, 5, 15],
'price': [300000, 400000, 250000, 500000, 350000] # Target variable for feature importance
}
df = pd.DataFrame(data)

# Correlation Analysis
correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Visualize correlation matrix (optional)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Feature Importance (using Random Forest)
features = df.drop('price', axis=1)
target = df['price']
model = RandomForestRegressor(random_state=42)
model.fit(features, target)
feature_importances = model.feature_importances_

# Print feature importances
print("\nFeature Importances:")
for feature, importance in zip(features.columns, feature_importances):
  print(f"{feature}: {importance:.4f}")
# Dimensionality Reduction (PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features) #scale only features

pca = PCA(n_components=3) #adjust based on explained variance.
reduced_data = pca.fit_transform(scaled_data)
explained_variance = pca.explained_variance_ratio_

print("\nExplained Variance Ratio:", explained_variance)

# Retained and Excluded Dimensions (based on PCA)
retained_dimensions = ["PC1", "PC2", "PC3"] #adjust based on n_components
excluded_dimensions = [col for col in features.columns]

# Print the names of the original columns.
print("\nOriginal columns:", features.columns.tolist())

# Create a DataFrame for the reduced data
reduced_df = pd.DataFrame(reduced_data, columns=retained_dimensions)

# Print reduced dataframe.
print("\nReduced DataFrame (PCA):\n", reduced_df)

# Print the retained and excluded dimensions.
print("\nRetained Dimensions (PCA):", retained_dimensions)

print("\nExcluded Dimensions (PCA):", excluded_dimensions)

print("\n loadings =",pca.components_)

# In a real scenario, you would determine which original dimensions were most responsible for the components.
# This requires looking at the components themselves (pca.components_)
# and interpreting their loadings.
```
